```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime
import pandas as pd
```

```python
df = pd.DataFrame(columns=['name', 'link', 'date', 'tags', 'author'])

page = 1
while True:
    resp = requests.get(f'https://hromadske.ua/news/page={page}')
    soup = BeautifulSoup(resp.text, "html.parser")
    for i in soup.findAll('article'): 
        name = i.find('a').find('h3').text
        link = i.find('a')['href']
        date = datetime.fromisoformat(i.find('time')['datetime'])
        article_soup = BeautifulSoup(requests.get(link).text, "html.parser")
        try:
            tags = [tag.text for tag in article_soup.findAll('a', 'c-button-outline c-button-outline--primary')]
            author = article_soup.find('a', 'c-post-author__name').text
            new_row = pd.DataFrame({'name': [name], 'link': [link], 'date': [date], 'tags': [tags], 'author': [author]})  
            df = pd.concat([df, new_row], ignore_index=True)
        except Exception as err:
                print(name, err)    
        page += 1
    print('-'*50,'size is:', df.shape[0])
    if df.shape[0] > 1000: break
```

```python
df.to_csv('hromadske.csv', index=False, encoding="utf-8")
```
